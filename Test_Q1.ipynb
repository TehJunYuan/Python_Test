```python
import random

# List of words for the game
words = ['PYTHON', 'PROGRAMMING']

# Randomly select a word from the list
word = random.choice(words).upper()
word_completion = ['_'] * len(word)  # List to hold the guessed state of the word
guessed_letters = []  # List to keep track of guessed letters
incorrect_guesses = 0  # Counter for incorrect guesses
max_incorrect_guesses = 6  # Maximum number of incorrect guesses allowed
guessed = False  # Flag to check if the word has been guessed

# Hangman stages
stages = [
    """
       ------
       |    |
       |    
       |    
       |    
       |   
       -
    """,
    """
       ------
       |    |
       |    O
       |    
       |    
       |   
       -
    """,
    """
       ------
       |    |
       |    O
       |    |
       |    
       |   
       -
    """,
    """
       ------
       |    |
       |    O
       |   \\|
       |    
       |   
       -
    """,
    """
       ------
       |    |
       |    O
       |   \\|/
       |    
       |   
       -
    """,
    """
       ------
       |    |
       |    O
       |   \\|/
       |    |
       |   / 
       -
    """,
    """
       ------
       |    |
       |    O
       |   \\|/
       |    |
       |   / \\
       -
    """
]

print("Let's play Hangman!")
print(stages[incorrect_guesses])
print(' '.join(word_completion))

while not guessed and incorrect_guesses < max_incorrect_guesses:
    guess = input('Guess a letter: ').upper()

    if len(guess) == 1 and guess.isalpha():
        print(f'Guess a letter: {guess}')

        if guess in guessed_letters:
            #check the guessed word
            print(f'You already guessed that letter.')
        elif guess not in word:
            #check the incorrect guessed word
            print(f'{guess} is not in the word.')
            incorrect_guesses += 1
            guessed_letters.append(guess)
        else:
            #check the correct guessed word
            print(f'Good job! {guess} is in the word!')
            guessed_letters.append(guess)
            for i in range(len(word)):
                if word[i] == guess:
                    word_completion[i] = guess
            if '_' not in word_completion:
                guessed = True
    else:
        #check invalid input. only alphabet is accepted
        print('Invalid input. Please enter a single letter.')

    print(stages[incorrect_guesses])
    print(' '.join(word_completion))
    print(f'Incorrect guesses: {incorrect_guesses}')

if guessed:
    print('Congratulations! You guessed the word correctly!')
else:
    print(f'Sorry, you ran out of guesses. The word was {word}. Better luck next time!')
```
