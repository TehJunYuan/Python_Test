```python
total_group_order = coffee.groupby(['product_category', 'product_type'])['transaction_qty'].sum().reset_index()
total_group_order.columns = ['category', 'product', 'total_order']

total_group_order
```

```python
total_group_order.head()
```

```python
total_group_order.tail()
```

```python
import matplotlib.pyplot as plt

plt.barh(total_group_order['product'], total_group_order['total_order'], color='blue')
plt.xlabel('Total order quantity')
plt.ylabel('Product type')
plt.title('Total orders by product type')

plt.show()
```
