```python
import pandas as pd

spotify_df = pd.read_csv("spotify.csv")

#extract year function
def extract_year(date_str):
    if isinstance(date_str, str):
        if len(date_str) == 4:  # Format: "YYYY"
            return int(date_str)
        elif len(date_str) >= 10:  # Format: "YYYY-MM-DD"
            return int(date_str[:4])
    return None

#extract the year and create new column for year
spotify_df['year'] = spotify_df['track_album_release_date'].apply(extract_year)
#only get the year == to 2017
df_2017 = spotify_df[spotify_df['year'] == 2017]

spotify_df
```

```python
#get data from only year 2017
df_2017 = spotify_df[spotify_df['year'] == 2017]
df_2017
```

```python
# Calculate the average values for the specified columns grouped by genre
genre_averages_2017 = df_2017.groupby('playlist_genre').agg({
    'loudness': 'mean',
    'mode': 'mean',
    'speechiness': 'mean',
    'acousticness': 'mean',
    'instrumentalness': 'mean',
    'liveness': 'mean',
    'valence': 'mean',
    'tempo': 'mean',
    'duration_ms': 'mean'
}).reset_index()
```

```python
# Convert duration from milliseconds to minutes
genre_averages_2017['duration_min'] = genre_averages_2017['duration_ms'] / 60000
average_metrics = genre_averages_2017.drop(columns='duration_ms')

output_path = 'average_metrics_2017.csv'
average_metrics.to_csv(output_path, index=False)
```

```python
# Investigating the relationship between danceability and energy
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_2017, x='danceability', y='energy', hue='playlist_genre', alpha=0.6)
plt.title('Danceability vs Energy')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.show()
```

```python
# Investigate correlation with other factors
correlation_matrix = df_2017[['danceability', 'energy', 'loudness', 'speechiness', 
                              'acousticness', 'instrumentalness', 'liveness', 
                              'valence', 'tempo']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix (2017)')
plt.show()
```
